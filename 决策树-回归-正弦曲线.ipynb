```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
rng = np.random.RandomState(1) #随机数种子
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
'''加噪声'''
y[::5] += 3 * (0.5 - rng.rand(16)) #y中每5个取出来一个，加上一个随机数（这个随机数是由-0.5到0.5之间的随机数组成的数组）
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.show()
```

```python
regr1 = DecisionTreeRegressor(max_depth=2)
regr2 = DecisionTreeRegressor(max_depth=5)
regr1.fit(X, y)
regr2.fit(X, y)
```

```python
X_test = np.arange(0, 5, 0.01)[:, np.newaxis]
y_1 = regr1.predict(X_test)
y_2 = regr2.predict(X_test)

#np.arange(0开始点, 5结束点, 0.01步长)[:, np.newaxis]增维度，变成二维
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")#原数据
plt.plot(X_test, y_1, color="cornflowerblue", label="max_depth=2", linewidth=2)#两条线的回归结果
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

#### 绿色的线更加容易收到噪声的影响, 过拟合
#### 蓝色的线更加稳定, 泛化能力强
