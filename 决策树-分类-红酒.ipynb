```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
wine
```

```python
wine.data
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
import pandas as pd
# 将 wine.data 和 wine.target 转换为 DataFrame 并合并
data_df = pd.DataFrame(wine.data, columns=wine.feature_names)
target_df = pd.DataFrame(wine.target, columns=['target'])
wine_df = pd.concat([data_df, target_df], axis=1)
wine_df
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(wine.data,wine.target,test_size=0.3)#划分训练集和测试集
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy',
                                  random_state=30,
                                  splitter='random',
                                  max_depth=5,
                                  )#进行决策树分类
clf = clf.fit(xtrain, ytrain)#fit相当于一个接口，引入训练集
score = clf.score(xtest, ytest)#测试集检测accuracy，对模型打分
score
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf,
                              feature_names=wine.feature_names,  #特征名
                              class_names=wine.target_names,  #类别名
                              filled=True, rounded=True)  #填充颜色
graph = graphviz.Source(dot_data) 
graph
```

```python
import matplotlib.pyplot as plt
test=[]
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1,
                                      criterion='entropy',
                                      random_state=30,
                                      splitter='random'
                                      )
    clf.fit(xtrain,ytrain)
    score=clf.score(xtest,ytest)
    test.append(score)
plt.plot(range(1,11),test,'o-')
plt.title('max_depth')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.show()

#剪枝参数
#min_samples_split: 
#min_samples_leaf: 
#min_weight_fraction_leaf: 
#max_features: 
#max_leaf_nodes: 
#min_impurity_decrease: 
#min_impurity_split: 
```
